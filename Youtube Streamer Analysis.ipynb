```python
df = pd.read_csv(file_path)  # Make sure the file path is correct
df.head(10)  # Preview the first 5 rows
```

```python
##Next, I proceed to clean the data by addressing missing values, renaming columns where necessary, and ensuring that each column is converted to the correct data type.
# Rename column 'suscribers' to 'subscribers'
df.rename(columns={'Suscribers': 'Subscribers'}, inplace=True)

# Convert 'Subscribers', 'Visits', 'Likes', and 'Comments' columns to numeric 
df['Subscribers'] = pd.to_numeric(df['Subscribers'], errors='coerce')
df['Visits'] = pd.to_numeric(df['Visits'], errors='coerce')
df['Likes'] = pd.to_numeric(df['Likes'], errors='coerce')
df['Comments'] = pd.to_numeric(df['Comments'], errors='coerce')

# Drop rows with missing values
df.dropna(inplace=True)

# Check for duplicates and remove them
df.drop_duplicates(inplace=True)
print(df.head())
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram of subscriber counts
plt.figure(figsize=(10, 6))
sns.histplot(df['Subscribers'], bins=30, kde=True, color='blue')

# Adding labels and title
plt.title('Distribution of Subscriber Counts Across All Streamers', fontsize=16)
plt.xlabel('Subscriber Count', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Sample data for demonstration 
top_categories = df.groupby('Categories')['Subscribers'].sum().sort_values(ascending=False).head(10)

# Creating a pie chart for top categories by subscriber count
plt.figure(figsize=(8, 8))
plt.pie(top_categories.values, 
        labels=top_categories.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=plt.cm.Paired.colors)

# Adding title
plt.title('Top Categories by Subscriber Count', fontsize=16)

# Displaying the chart
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot for the distribution of subscriber counts
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Subscribers'], color='lightblue')

# Adding labels and title
plt.title('Box Plot of Subscriber Counts Across All Streamers', fontsize=16)
plt.xlabel('Subscriber Count', fontsize=12)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by country and get the count of streamers, then sort and take the top 10
top_countries = df['Country'].value_counts().head(10)

# Create a bar chart for the top 10 countries with the most streamers
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='coolwarm')

# Add labels and title
plt.title('Top 10 Countries with the Most Streamers', fontsize=16)
plt.xlabel('Number of Streamers', fontsize=12)
plt.ylabel('Country', fontsize=12)

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average likes and comments per visit
df['likes_per_visit'] = df['Likes'] / df['Visits']
df['comments_per_visit'] = df['Comments'] / df['Visits']

# Create a DataFrame to hold the calculated averages
engagement_metrics = pd.DataFrame({
    'Average Likes per Visit': [df['likes_per_visit'].mean()],
    'Average Comments per Visit': [df['comments_per_visit'].mean()]
})

# Transpose the DataFrame for easier plotting
engagement_metrics = engagement_metrics.T
engagement_metrics.columns = ['Value']

# Visualize the average likes and comments per visit using a bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=engagement_metrics.index, y=engagement_metrics['Value'], palette='viridis')

# Add labels and title
plt.title('Average Likes and Comments per Visit (Audience Engagement)', fontsize=16)
plt.xlabel('Engagement Metric', fontsize=12)
plt.ylabel('Average Value per Visit', fontsize=12)

# Display the chart
plt.show()
```

```python
## Using heatmap to show the relationship between subscibers, visits, likes and comment
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(df[['Subscribers', 'Visits', 'Likes', 'Comments']].corr(), dtype=bool))

# Plot the correlation heatmap with the mask
plt.figure(figsize=(10, 8))
sns.heatmap(df[['Subscribers', 'Visits', 'Likes', 'Comments']].corr(), annot=True, cmap='coolwarm', linewidths=0.5, mask=mask)

# Add title
plt.title('Correlation Heatmap', fontsize=16)

# Display the plot
plt.show()
```
